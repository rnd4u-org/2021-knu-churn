```python
import pandas
import numpy as np
import seaborn as sns

from matplotlib import pyplot as plt

from sklearn.model_selection import train_test_split 
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier  
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score


from sklearn.feature_selection import SelectKBest, f_classif, chi2,  SelectFromModel
```

```python
churners = pandas.read_csv("BankChurnersUpdated.csv")
churners_original = pandas.read_csv("BankChurners.csv")

churners
```

```python
churners_X = churners.drop(["Attrition_Flag", 
                            "Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1", 
                            "Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2"], 
                            axis=1)
columns = pandas.DataFrame(churners_original.columns[2:21])

X = churners_X.values
y = churners["Attrition_Flag"].values

X_train, X_test, y_train, y_test = train_test_split( 
              X, y, test_size = 0.3)

churners_X
```

```python
classifier =  RandomForestClassifier()
classifier = classifier.fit(X_train, y_train)
predicted = classifier.predict(X_test)

before_feature_selection = 'Confusion Matrix : \n' + str(confusion_matrix(y_test, predicted)) + '\n Accuracy Score : ' + \
                str(accuracy_score(y_test, predicted)) + '\n f1_score: ' + str(f1_score(y_test, predicted)) + \
               "\n recall_score: " + str(recall_score(y_test,predicted)) + "\n precision_score: " + str(precision_score(y_test,predicted)) + "\n"
```

```python
best_features_selector = SelectKBest(chi2)
fit = best_features_selector.fit(X, y)
scores = pandas.DataFrame(fit.scores_)

features_scores = pandas.concat([columns, scores], axis=1)
features_scores.columns = ('Features', 'Score')
best_features = []

for feature in features_scores.values:
    if feature[1] > 2:
        best_features.append(feature[0])

print("Best features are: ", best_features)
features_scores
```

```python
X_best = churners[best_features]
X_train, X_test, y_train, y_test = train_test_split(X_best, y, test_size=0.3)

print("Results before model upgrade: \n")
print(before_feature_selection)

print("Results after model upgrade: \n ")
classifier =  RandomForestClassifier()
classifier = classifier.fit(X_train, y_train)
predicted = classifier.predict(X_test)

print ('Confusion Matrix :') 
print(confusion_matrix(y_test, predicted)) 
print ('Accuracy Score :',accuracy_score(y_test, predicted)) 
print ('f1_score: ', f1_score(y_test, predicted)) 
print ("recall_score", recall_score(y_test,predicted))
print ("precision_score: ", precision_score(y_test,predicted))
```

```python
print(X.shape)

classifier = ExtraTreesClassifier(n_estimators=50)
classifier = classifier.fit(X, y)  

model = SelectFromModel(classifier, prefit=True)
X_new = model.transform(X)

print(X_new.shape)

print("Results before model upgrade: \n ")
print(before_feature_selection)

X_train, X_test, y_train, y_test = train_test_split( 
              X_new, y, test_size = 0.3)

classifier =  RandomForestClassifier()
classifier = classifier.fit(X_train, y_train)
predicted = classifier.predict(X_test)

print("Results after model upgrade: \n ")
print ('Confusion Matrix :') 
print(confusion_matrix(y_test, predicted)) 
print ('Accuracy Score :',accuracy_score(y_test, predicted)) 
print ('f1_score: ', f1_score(y_test, predicted)) 
print ("recall_score", recall_score(y_test,predicted))
print ("precision_score: ", precision_score(y_test,predicted))
```
