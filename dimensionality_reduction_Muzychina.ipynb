```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, TruncatedSVD, NMF


data = pd.read_csv('drive/MyDrive/Colab Notebooks/BankChurners_ML.csv')
x = data[data.columns[1:20]]
x
```

# Using 3 matrix factorization methods, project features on a subspace

```
proj1 = {}
proj1[2] = PCA(2).fit(x).transform(x)
proj1[3] = PCA(3).fit(x).transform(x)


proj2 = {}
proj2[2] = TruncatedSVD(2).fit(x).transform(x)
proj2[3] = TruncatedSVD(3).fit(x).transform(x)


proj3 = {}
proj3[2] = NMF(2).fit(x).transform(x)
proj3[3] = NMF(3).fit(x).transform(x)
```

# Using 3 methods, visualize feature space in 2D and 3D. Try to spot inner structure.

```
fig = plt.figure(figsize=(30, 12))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122, projection='3d')

ax1.scatter(x=proj1[2][:, 0], y=proj1[2][:, 1])
ax2.scatter(xs=proj1[3][:, 0], ys=proj1[3][:, 1], zs=proj1[3][:, 2])


fig = plt.figure(figsize=(30, 12))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122, projection='3d')

ax1.scatter(x=proj2[2][:, 0], y=proj2[2][:, 1])
ax2.scatter(xs=proj2[3][:, 0], ys=proj2[3][:, 1], zs=proj2[3][:, 2])


fig = plt.figure(figsize=(30, 12))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122, projection='3d')

ax1.scatter(x=proj3[2][:, 0], y=proj3[2][:, 1])
ax2.scatter(xs=proj3[3][:, 0], ys=proj3[3][:, 1], zs=proj3[3][:, 2])

plt.show()
```
